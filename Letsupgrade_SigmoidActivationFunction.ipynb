#Tanh vs sigmoid


```python
x = np.linspace(-10,10,100)

#sigmoid activation function
sigmoid = 1/(1+np.exp(-x))
sigmoid_derivative = sigmoid *(1-sigmoid)

#tanh activation function
tanh = (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
tanh_derivative = 1-np.square(tanh)
```

```python
df = pd.DataFrame(dict(
    x=x,
    sigmoid = sigmoid,
    sigmoid_derivative = sigmoid_derivative,
    tanh = tanh,
    tanh_derivative = tanh_derivative
))
```

```python
fig = df.plot.scatter(x='x',y='sigmoid')
fig.show()
```

```python
fig = df.plot.scatter(x='x',y='tanh')
fig.show()
```

#Sigmoid and Tanh in Forward Propagation

```python
fig = df.plot.scatter(x='x',y=['sigmoid','tanh'])
fig.show()
```

#Sigmoid and tanh in backward propagation

```python
fig  = df.plot.scatter(x='x',y=['sigmoid_derivative','tanh_derivative'])
fig.show()
```

```python
def Relu(x):
  return x * (x>0)
```

#Derivative of relu

```python
def DRelu(x):
  return 1*(x>0)
```

```python
df = pd.DataFrame(dict(
    x=x,
    relu =Relu(x),
    Drelu = DRelu(x)
))
```

```python
fig = df.plot.scatter(x='x',y = 'Drelu')
fig.show()
```
